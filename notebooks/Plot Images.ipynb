# Plotting Cell Image Examples

```python
import os
```

```python
# Prepare paths
cwd = os.getcwd()  # current working folder
pardir = os.path.dirname(cwd)   # parent folder

all_folder = os.path.join(pardir, 'data', 'all_folder')
hem_folder = os.path.join(pardir, 'data', 'hem_folder')
```

# Load images into arrays

```python
import numpy as np
from skimage import io
from PIL import Image
```

```python
# read all images from disk and store them in an array
all_list=[]
all_paths = os.listdir(all_folder)
for i, image_name in enumerate(all_paths):
    if(image_name.split('.')[1] == 'bmp'):
        image = io.imread(all_folder+'/'+image_name)
        image = Image.fromarray(image, 'RGB')
        # image = image.resize((150,150)) #150x150
        all_list.append(np.array(image))
all_array = np.array(all_list)

del all_list
```

```python
all_array.shape
```

```python
# read hem images from disk and store them in an array
hem_list=[]
hem_paths = os.listdir(hem_folder)
for i, image_name in enumerate(hem_paths):
    if(image_name.split('.')[1] == 'bmp'):
        image = io.imread(hem_folder+'/'+image_name)
        image = Image.fromarray(image, 'RGB')
        #image = image.resize((210,210)) #150x150
        hem_list.append(np.array(image))
hem_array = np.array(hem_list)

del hem_list
```

```python
hem_array.shape
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(1, 3, figsize=(6, 2.4))
for i, ax in enumerate(axes.flat):
    if i in range(6):
        ax.imshow(all_array[i],cmap='gray', alpha= 0.95)
        ax.axis('off')
fig.suptitle('Leukemia Cells (ALL)', fontsize=16, fontweight='bold')
fig.tight_layout()

fig, axes = plt.subplots(1, 3, figsize=(6, 2.4))
for i, ax in enumerate(axes.flat):
    if i in range(6):
        ax.imshow(hem_array[i],cmap='gray', alpha=0.94)
        ax.axis('off')
fig.suptitle('Healthy Cells (HEM)', fontsize=16, fontweight='bold')
fig.tight_layout()

plt.show()
```
